```python
import requests
import json
import base64
import tensorflow as tf
import io
from PIL import Image
```

```python
image_data=tf.io.read_file('object-detection-SMALL_2_files\\000000150805.jpg')
image_data
```

```python
# Encode the tensor as base64
encoded_tensor = tf.io.encode_base64(image_data)

# Decode the byte string to UTF-8 string
base64_string = encoded_tensor.numpy().decode('utf-8')

print(base64_string)
```

```python
# Convert the base64 string to a byte string
byte_string = base64_string.encode('utf-8')

# Decode the byte string to a TensorFlow tensor
decoded_tensor = tf.io.decode_base64(byte_string)
decoded_tensor
```
